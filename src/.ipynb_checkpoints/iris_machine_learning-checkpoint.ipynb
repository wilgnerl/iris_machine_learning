* **Objetivo prever a especie de planta a partir das caracteristicas**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
df = pd.read_csv('../data/iris.csv')
df.head()
```

```python
df.Species.unique()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
columns_name = list(df.columns)
del columns_name[0]
del columns_name[4]

columns_name
```

```python
names = dict()
traducao = ['Altura_Sepala', 'Largura_Sepala', 'Altura_Petala', 'Largura_Petala']

for index,name in enumerate(traducao):
    names[columns_name[index]]=name
```

```python
names
```

```python
df.rename(names, axis='columns', inplace=True)
df.head()
```

```python
especies = list(df.Species.unique())
especies

dummies = dict()

for index, name in enumerate(especies):
    dummies[name]=index
    
dummies
```

```python
df['Species_dummy'] = df.Species
df['Species_dummy'] = df['Species_dummy'].map(dummies)
```

```python
fig = plt.figure(figsize=(20,15))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

subplots = [ax1,ax2,ax3,ax4]

for index, name in enumerate(traducao):
    sns.distplot(df[name], ax=subplots[index], bins=12,
                 hist_kws={'density':True, "linewidth": 1,
                           "alpha": 1, "color": "g", 'edgecolor':'white'}, kde=False)
    subplots[index].set_title(name, fontsize = 18, fontweight='bold')
```

```python
nomes_plot = list(['Id','Species' ])+traducao
nomes_plot
```

```python
sns.pairplot(df[nomes_plot], hue='Species')
```

```python
from sklearn.model_selection import train_test_split
X = df[traducao]
y = df['Species_dummy']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=33)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth=3, random_state=0, n_estimators=1000)
clf.fit(X_train, y_train)
```

```python
resultado = clf.predict(X_test)
type(resultado)
```

```python
from sklearn.metrics import accuracy_score

y_pred = resultado
y_true = np.array(y_test)
```

```python
accuracy_score(y_true, y_pred, normalize=True)
```

```python
pd.Series(resultado).value_counts()
```

```python
from sklearn.metrics import confusion_matrix

matriz = confusion_matrix(y_true, y_pred)

sns.heatmap(matriz, annot=True)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_true, y_pred, target_names = list(dummies.keys())))
```
